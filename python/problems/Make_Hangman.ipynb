## 답안 확인 하기

> 행맨 게임을 위한 코드를 작성해봅시다. 
>
> `isanswer(answer, letters)`는 정답 문자열(answer)과 추측 알파벳의 리스트(letters)를 받아서, True/False 결과를 반환합니다.

```python
# 여기에 코드를 작성하세요.
def isanswer(answer, letters):
    if ''.join(letters)==answer:
        return True
    else :
        return False
        
        
```

```python
isanswer('apple', ['a','p','p','e']) #=> False
```

## 현재 상태 출력하기

> 행맨 게임을 위한 코드를 작성해봅시다. 
>
> `status(answer, letters)`는 정답 문자열(answer)과 추측 알파벳의 리스트(letters)를 받아서, 현재 결과를 반환합니다.

```python
# replace
def status(answer, letters, lst):
    for j,k in enumerate(answer):
        if k in letters:
            lst[j]=k
    return lst
```

```python
status('apple', ['a', 'p', 'l']) #=> 'appl_'
```

## 행맨 로직 만들기 - 자유코딩!

> 위의 코드를 활용하여 `hangman(answer)`를 만들어봅시다!
>
> 사용자가 답을 맞출 때까지 진행하며, 8번내로 정답을 맞춰야 합니다.


```python
def hangman():
    answer=call()
    answerlist=['']
    mylst=[]
    lst=['_' for i in answer]
    if len(answer)>18:
        life=14
    elif len(answer)>15:
        life=13
    elif len(answer)>10:
        life=12
    elif len(answer)>6:
        life=11
    else:
        life=10
    count=0
    while True :  
        a = input('알파벳을 입력하세요')
        if a in mylst:
            print('이미 입력한 문자입니다. 다시입력하세요')
            continue
        else:
            mylst.append(a)
        lst=status(answer, mylst,lst[:])
        
        if a in lst:
            print(f'정답에 가까워 지셨습니다.\n{lst}\n')
        else:
            life-=1
            print(f'입력하신 알파벳은 정답안에 존재하지 않습니다. Life : {life}\n')
        if isanswer(answer,lst):
            print('정답입니다!')
            break
        if life<1:
            print(f'Game Over\n정답은 {answer}입니다.')
            break
        print(f'현재까지 입력된 문자 : {set(mylst)}')  
        
```
